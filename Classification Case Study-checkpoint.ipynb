```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from datetime import datetime

# --- IMPORTANT: CHANGE THIS TO YOUR ACTUAL FILE PATH ---
FILE_PATH = r"C:\Users\naush\Downloads\advertising (1).csv"
# --------------------------------------------------------

try:
    df = pd.read_csv(FILE_PATH)
except FileNotFoundError:
    print(f"ERROR: File not found at {FILE_PATH}. Please check the path and filename.")
    # Stop execution if file is not found
    raise 

print("--- Initial Data Inspection (df.head() and df.info()) ---")
print(df.head())
df.info()

# 1. Handling Missing Values and Duplicates
print("\n--- Checking for Missing Values and Duplicates ---")
print("Missing values per column:\n", df.isnull().sum())
print(f"Number of duplicate rows: {df.duplicated().sum()}")

# Assuming simple removal for illustration, but imputation might be better depending on the data
df.dropna(inplace=True) 
df.drop_duplicates(inplace=True)

# 2. Outlier Handling (Simple capping for numerical features)
numerical_cols = ['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage']

# Function for simple IQR-based outlier capping
def cap_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return np.where(series > upper_bound, upper_bound, np.where(series < lower_bound, lower_bound, series))

for col in numerical_cols:
    df[col] = cap_outliers(df[col])

print("\nData Cleaning Complete. Remaining rows:", len(df))
```

```python
# 3. Exploratory Data Analysis (EDA)
print("\n--- Exploratory Data Analysis (EDA) ---")

# a) Target Variable Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Clicked on Ad', data=df)
plt.title('Distribution of the Target Variable (Clicked on Ad)')
plt.show()

# b) Relationship between Age/Income/Usage and Click
plt.figure(figsize=(18, 5))
for i, col in enumerate(numerical_cols):
    plt.subplot(1, 4, i + 1)
    sns.boxplot(x='Clicked on Ad', y=col, data=df)
    plt.title(f'{col} vs. Clicked on Ad')
plt.tight_layout()
plt.show() 


# c) Feature Engineering from Timestamp
# Extract features (e.g., Hour of Day)
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Hour'] = df['Timestamp'].dt.hour

# d) Click Rate by Hour
hourly_click_rate = df.groupby('Hour')['Clicked on Ad'].mean().reset_index()
plt.figure(figsize=(8, 5))
sns.barplot(x='Hour', y='Clicked on Ad', data=hourly_click_rate)
plt.title('Click Rate by Hour of Day')
plt.ylabel('Average Click Rate')
plt.show()
```

```python
# 4. Feature Engineering and Data Preprocessing

# Define features (X) and target (y)
X = df.drop(['Clicked on Ad', 'Timestamp', 'Ad Topic Line'], axis=1) # Dropping Ad Topic Line for simplicity, and Timestamp after extracting 'Hour'
y = df['Clicked on Ad']

# Define which columns are numerical and which are categorical
numerical_features = ['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Hour']
categorical_features = ['City', 'Country', 'Male'] # Male is already binary, but will be treated here for robustness

# Create a preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        # Standardize numerical features
        ('num', StandardScaler(), numerical_features),
        # One-Hot Encode categorical features
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough' # Keep other columns if any, though none are expected here
)

# 5. Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
print(f"\nData split complete. Training size: {len(X_train)}, Testing size: {len(X_test)}")

# 6. Build Classification Model (Logistic Regression)
# Create the full modeling pipeline (Preprocessing + Logistic Regression)
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=42, solver='liblinear'))
])

print("\n--- Training Logistic Regression Model ---")
model_pipeline.fit(X_train, y_train)
```

```python
# 7. Model Evaluation
print("\n--- Model Evaluation ---")

# Predict on the test set
y_pred = model_pipeline.predict(X_test)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# Detailed classification report
print("\n--- Classification Report ---")
print(classification_report(y_test, y_pred))

# Confusion Matrix visualization
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['No Click (0)', 'Click (1)'], 
            yticklabels=['No Click (0)', 'Click (1)'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()


# 8. Reporting and Visualization (Model Insights)
# Extracting feature importance (for logistic regression, coefficients)
# This is complex due to the pipeline, but here's a simplified way for discussion:
print("\n--- Key Findings for Reporting ---")
print(f"The model successfully predicts customer clicks with an F1-Score of {f1:.2f}.")
print("Insights from EDA (e.g., specific age groups or internet usage patterns) and feature coefficients (from the model) will form the core of your strategic recommendations for targeted advertising.")
```
